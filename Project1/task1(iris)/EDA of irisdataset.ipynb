## Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the Iris dataset

```python
file_path = "C:/Users/Vaibhav/Desktop/Project1/Iris.csv" 
df = pd.read_csv(file_path)
```

## Display the first few rows of the dataset

```python
print(df.head())
```

## Get dataset information

```python
print(df.info())
```

## Check for missing values,if any

```python
print(df.isnull().sum())
```

## Descriptive statistics
#### This prints descriptive statistics (mean, standard deviation, minimum, maximum, etc.) for each numeric column in the dataset

```python
print(df.describe())
```

## Histograms for each feature
#### The histograms help you understand the spread and concentration of values for each numeric feature in the Iris dataset.

```python
plt.figure(figsize=(10, 8))
for i, column in enumerate(df.columns[1:-1]):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
```

## Box plots for each feature
#### It helps to analyse the position of the median, the spread of the interquartile range, and the presence of outliers can be compared across species for each feature and help to compare different species.

```python
plt.figure(figsize=(10, 8))
for i, column in enumerate(df.columns[1:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='Species', y=column, data=df)
    plt.title(f'Box plot of {column} by Species')
    
plt.show()
```

## Correlation matrix
#### The heatmap is a valuable tool for identifying patterns and relationships in the data. High positive correlations are typically indicated by warmer colors, high negative correlations by cooler colors, and weaker correlations by colors closer to zero. 

```python
correlation_matrix = df[df.columns[1:-1]].corr()
```

## Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title("Correlation Matrix")
plt.show()
```
